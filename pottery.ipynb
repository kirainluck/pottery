## Converting Adobe illustrator format to png

```
import fitz
pdffile = "/content/drive/My Drive/For Kyrill/For Kyrill/Cup Illustrations/p3084-15.ai"
doc = fitz.open(pdffile)
page = doc.loadPage(0)  # number of page
pix = page.getPixmap()
output = "/content/drive/My Drive/For Kyrill/For Kyrill/Cup Illustrations/outfile.png"
pix.writePNG(output)
```

```
import os

folder = "/content/drive/My Drive/For Kyrill/For Kyrill/Cup Illustrations/"
for filename in os.listdir(folder):
  doc = fitz.open(os.path.join(folder,filename))
  page = doc.loadPage(0)  # number of page
  pix = page.getPixmap()
  newfolder = "/content/drive/My Drive/For Kyrill/For Kyrill/pngs"
  pix.writePNG(os.path.join(newfolder,f"{filename}.png"))
```

All files saved to '/content/drive/My Drive/For Kyrill/For Kyrill/pngs'

## Displaying all pottery pngs

```
import matplotlib.cbook as cbook
from matplotlib import pyplot as plt
folder = "/content/drive/My Drive/For Kyrill/For Kyrill/pngs"
i = 1
columns = 15
rows = 15
fig=plt.figure(figsize=(15, 15))
for filename in os.listdir(folder):
  with cbook.get_sample_data(os.path.join(folder,filename)) as image_file:
    img = plt.imread(image_file)
    fig.add_subplot(rows, columns, i)
    plt.imshow(img)
    i+=1
plt.show()
```
